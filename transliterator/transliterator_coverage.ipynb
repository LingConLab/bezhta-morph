Make sure that `hfst-lookup bezhta.tr.analyzer.hfst` is in the same directory

```python
def check_word(word):
  output = os.popen(f"echo {word} | hfst-lookup bezhta.tr.analyzer.hfst").read().rstrip('\n\n')
  parses = []
  for el in output.split('\n'):
      parses.append(':'.join(el.split('\t')))

  return parses
```

```python
from collections import defaultdict
```

```python
def check_tr_coverage(path):
   with open(path, 'r', encoding='utf-8') as file:
    un = []
    total = 0.0
    tokens_recognised = 0.0
    unique = 0.0
    types_recognised = 0.0
    unrecog = defaultdict(int)
    lines = file.readlines()
    for l in lines:
        w = l.strip()
        parses = check_word(w)
        if '?' not in parses[0]:
          tokens_recognised += 1
          if w not in un:
            types_recognised +=1
        else:
          unrecog[w] += 1
        if w not in un:
            un.append(w)
        total +=1
    print('tokens: ', total)
    print('types: ', len(un))
    print('tokens recognised: ', tokens_recognised)
    print('token coverage: ', tokens_recognised / total)
    print('types recognised: ', types_recognised)
    print('type coverage: ', types_recognised / len(un))
    return unrecog
```

```python
check_tr_coverage('valpal_examples.txt')
```
